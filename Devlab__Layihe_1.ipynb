```python
import pandas as pd
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['ORDERDATE'].dtype
```

```python
df['ADDRESSLINE2'] = df['ADDRESSLINE2'].fillna('Unknown')
df['STATE'] = df['STATE'].fillna('Unknown')
df['TERRITORY'] = df['TERRITORY'].fillna('Unknown')
df['POSTALCODE'] = df['POSTALCODE'].fillna('Unknown')

df.isnull().sum()
```

```python
duplicate_count = df.duplicated().sum()
print(f"Dublikat sətir sayı: {duplicate_count}")
```

```python
df.describe()
```

```python
print(df['STATUS'].value_counts())
print(df['YEAR_ID'].value_counts())
print(df['PRODUCTLINE'].value_counts())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
df['SALES'].hist(bins=30, color='steelblue', edgecolor='black')
plt.title('Satış Məbləğlərinin Paylanması (SALES Histoqramı)')
plt.xlabel('Satış Məbləği ($)')
plt.ylabel('Tezlik')
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['YEAR_ID'].value_counts().sort_index().plot(kind='bar', color='coral')
plt.title('İllər Üzrə Sifariş Sayı')
plt.xlabel('İl')
plt.ylabel('Sifariş Sayı')
plt.xticks(rotation=0)
plt.show()
```
